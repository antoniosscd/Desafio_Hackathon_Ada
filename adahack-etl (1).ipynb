---
A base de dados que temos em m√£os n√£o oferece com exatid√£o a quantidade atual de funcionarios, vamos deduzir emt√£o que a quantidade √© proporcional, levando em consider√ß√£o o valor relativo de cada grupo. Portanto temos:

```python
crosstab = pd.crosstab([base_dados['Genero'], base_dados['Ra√ßa']], base_dados['Senioridade'])

# Plotar o gr√°fico de barras agrupadas com dist√¢ncia entre as barras aumentada
ax = crosstab.plot(kind='bar', figsize=(25, 12), width=0.6)  # Ajuste a largura das barras conforme necess√°rio
plt.title('Distribui√ß√£o de G√™nero, Ra√ßa e Senioridade')
plt.xlabel('G√™nero e Ra√ßa')
plt.ylabel('Quantidade')
plt.legend(title='Senioridade')

# Adicionar as quantidades nas barras
for i in ax.patches:
    ax.text(i.get_x() + i.get_width() / 2, i.get_height() + 0.05, str(i.get_height()), ha='center', va='bottom')

plt.show()
```

```python
contagem_genero = base_dados['Genero'].value_counts()
contagem_genero = contagem_genero.reindex(['Masc', 'Fem'])

# Crie um gr√°fico de barras
plt.bar(contagem_genero.index, contagem_genero.values)

# Adicione r√≥tulos e t√≠tulo
plt.xlabel('G√™nero')
plt.ylabel('Quantidade')
plt.title('Contagem de G√™neros na Base de Dados')

# Adicione os r√≥tulos das barras
for i, value in enumerate(contagem_genero.values):
    plt.text(i, value + 50, str(value), ha='center', va='bottom')

# Exiba o gr√°fico
plt.show()
```

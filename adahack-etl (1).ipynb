```python
#Gerando a meta de Genero.
total_funcionarios = 850
tempo_mes = 18
genero_meta = {}

# tabela com dados censo2022
ibge = {
    'Genero':['Masc','Fem'],
    'Total%':[0.4851,0.5148]
         }
ibge = pd.DataFrame(ibge)

#tabela com a meta.
genero_meta= {
    'Genero':['Masc','Fem'],
    'Atual': [round(total_funcionarios * ((base_dados['Genero']=='Masc').sum())/len(base_dados['Genero'])), round(total_funcionarios * ((base_dados['Genero']=='Fem').sum())/len(base_dados['Genero']))],
    'IBGE%': [48.51,51.48],
    'Meta 2026': [round(total_funcionarios * ibge['Total%'][0]), round(total_funcionarios * ibge['Total%'][1])],
    'Meta mes' :[round(total_funcionarios * ibge['Total%'][0] / tempo_mes),
                 round(total_funcionarios * ibge['Total%'][1] / tempo_mes)]
}
genero_meta = pd.DataFrame(genero_meta)

print(genero_meta)
print(ibge)
```

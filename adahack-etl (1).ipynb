A inten√ß√£o √© equalizar os g√™neros. Vamos criar uma tabela contendo os dados atuais, os dados do censo de 2022 e as metas mensais at√© 2026 de cada grupo, e gerar um gr√°fico com esses dados.

---

```python
#Gerando as metas para Genero.
total_funcionarios = 850 #meta para o quardro ate 2026
quadro_atual = 425 #Quadro atual 
tempo_mes = 18 #Quantos meses faltam para 2026
genero_meta = {}  #Dicionario Vazio para armazenar as tabelas.

# Tabela com dados do censo de 2022. A chave 'Total%' mapeia o valor relativo de cada grupo dentro da 
# popula√ß√£o.
ibge = {
    'Genero':['Masc','Fem'],
    'Total%':[0.4851,0.5148]
         } 
ibge = pd.DataFrame(ibge)

#Descreve as metas mensais e o total at√© 2026 de cada grupo.

genero_meta= {
    'Genero':['Masc','Fem'],
    #Calculamos a propor√ß√£o atual de cada grupo com base nos valores relativos nos dados hist√≥ricos, 
    # seguindo a f√≥rmula total_do_grupo / total_de_registros. Essa propor√ß√£o √© obtida dividindo o total 
    # de cada grupo pelo total de registros. Com esse c√°lculo, podemos deduzir a quantidade atual de 
    # cada grupo, j√° que os dados n√£o nos fornecem essa informa√ß√£o.  
    'Atual': [round(quadro_atual * ((base_dados['Genero']=='Masc').sum())/len(base_dados['Genero'])), 
              round(quadro_atual * ((base_dados['Genero']=='Fem').sum())/len(base_dados['Genero']))],
    #Descreve os valores relativos de cada grupo segundo ibge.
    'IBGE%': [48.51,51.48],
    #Calcula o total de 850 multiplicado pelo valor relativo do grupo no censo de 2022, obtendo a meta do grupo para 2026.
    'Meta 2026': [round(total_funcionarios * ibge['Total%'][0]),
                  round(total_funcionarios * ibge['Total%'][1])],
    #Calcula o total de 850 multiplicado pelo valor relativo do grupo no censo de 2022, obtendo a meta do grupo para cada m√™s.
    'Meta mes' :[round(total_funcionarios * ibge['Total%'][0] / tempo_mes),
                 round(total_funcionarios * ibge['Total%'][1] / tempo_mes)]
}
genero_meta = pd.DataFrame(genero_meta)

genero_meta
```

```python
plt.figure(figsize=(10, 6))
barWidth = 0.20


r1 = range(len(genero_meta))


plt.bar(r1, genero_meta['Atual'], color='c', width=barWidth, edgecolor='grey', label='Atual')
plt.bar([r + barWidth for r in r1], genero_meta['Meta 2026'], color='y', width=barWidth, edgecolor='grey', label='Meta 2026')
plt.bar([r + 2*barWidth for r in r1], genero_meta['Meta mes'], color='pink', width=barWidth, edgecolor='grey', label='Meta mes')


for i, value in enumerate(genero_meta['Atual']):
    plt.text(i, value + 5, str(value), color='black', ha='center')
for i, value in enumerate(genero_meta['Meta 2026']):
    plt.text(i + barWidth, value + 5, str(value), color='black', ha='center')
for i, value in enumerate(genero_meta['Meta mes']):
    plt.text(i + 2*barWidth, value + 5, str(value), color='black', ha='center')


plt.xlabel('G√™nero', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(genero_meta))], genero_meta['Genero'])
plt.ylabel('Valores')
plt.title('Compara√ß√£o entre Atual e Meta 2026 por G√™nero')
plt.legend()

plt.show()
```

```python
total_funcionarios = 850
tempo_mes = 18
quadro_atual = 425

atual = base_dados['Ra√ßa'].value_counts(normalize=True) * quadro_atual
atual = atual.reindex(['pardo', 'branco', 'preto', 'ind√≠gena', 'amarelo']).fillna(0).astype(int)

# C√°lculo dos valores de 'Meta 2026' e 'Meta mes'
meta_2026 = total_funcionarios * raca_ibge['% Populacao'] / 100
meta_mes = meta_2026 / tempo_mes

# Cria√ß√£o do DataFrame
raca_meta = pd.DataFrame({
    'Raca': ['pardo', 'branco', 'preto', 'ind√≠gena', 'amarelo'],
    'Atual': atual.values,
    'Meta 2026': meta_2026.round().astype(int),
    'Meta mes': meta_mes.round().astype(int)
})

raca_meta
```

```python
plt.figure(figsize=(10, 4))
barWidth = 0.20

r1 = range(len(raca_meta))

for i, col in enumerate(['Atual', 'Meta 2026', 'Meta mes']):
    plt.bar([r + i*barWidth for r in r1], raca_meta[col], color=['c', 'y', 'pink'][i],
            width=barWidth, edgecolor='grey', label=col)
    for j, value in enumerate(raca_meta[col]):
        plt.text(j + i*barWidth, value + 5, str(value), color='black', ha='center')

plt.xlabel('Ra√ßa', fontweight='bold')
plt.xticks([r + barWidth for r in r1], raca_meta['Raca'])
plt.ylabel('Valores')
plt.title('Quadro Atual VS Meta 2026')
plt.legend()

plt.show()
```
